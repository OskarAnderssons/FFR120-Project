```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from CSV files
df_1 = pd.read_csv('Data_1312/Fish_eaten_log_1_1312.csv')
df_05 = pd.read_csv('Data_1312/Fish_eaten_log_05_1312.csv')
df_025 = pd.read_csv('Data_1312/Fish_eaten_log_025_1312.csv')
df_15 = pd.read_csv('Data_1312/Fish_eaten_log_15_1312.csv')
max_lifetime = df_1['Time Elapsed'].max()

slope_15 = np.polyfit(df_15['Time Elapsed'], df_15['Fish Eaten'], 1)[0]
slope_1 = np.polyfit(df_1['Time Elapsed'], df_1['Fish Eaten'], 1)[0]
slope_05 = np.polyfit(df_05['Time Elapsed'], df_05['Fish Eaten'], 1)[0]
slope_025 = np.polyfit(df_025['Time Elapsed'], df_025['Fish Eaten'], 1)[0]

print('Slope 1.5: ', slope_15, 'Slope 1: ', slope_1, 'Slope 0.5: ', slope_05, 'Slope 0.25: ', slope_025)
# Create the scatter and line plot
plt.figure(figsize=(12, 6))

# Scatter plot of raw data
plt.scatter(df_1['Time Elapsed'], df_1['Fish Eaten'], color='blue', alpha=1, label='10 Raw Data')
plt.scatter(df_05['Time Elapsed'], df_05['Fish Eaten'], color='red', alpha=1, label='0.5 Raw Data')
plt.scatter(df_025['Time Elapsed'], df_025['Fish Eaten'], color='green', alpha=1, label='0.25 Raw Data')
plt.scatter(df_15['Time Elapsed'], df_15['Fish Eaten'], color='orange', alpha=1, label='1.5 Raw Data')
# Add titles and labels
plt.title('Fish Eaten vs Time', fontsize=16, fontweight='bold')
plt.xlabel('Time Elapsed', fontsize=14)
plt.ylabel('Total fish eaten', fontsize=14)

# Customize ticks and grid
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
# plt.xscale('log')
# plt.yscale('log')

# Add legend
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
