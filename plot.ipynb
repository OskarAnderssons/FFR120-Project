```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#200 fish, 15000 steps, 3 predators
# Load the data from a CSV file
df_10 = pd.read_csv('fish_data_10deltaT.csv')
df_15 = pd.read_csv('fish_data_15deltaT.csv')
df_5 = pd.read_csv('fish_data_5deltaT.csv')

df_10_filtered = df_10[df_10['Average Neighbors'] <= 100]
df_15_filtered = df_15[df_15['Average Neighbors'] <= 100]
df_5_filtered = df_5[df_5['Average Neighbors'] <= 100]

df_10_filtered = df_10_filtered[df_10_filtered['Lifetime'] > 1]
df_15_filtered = df_15_filtered[df_15_filtered['Lifetime'] > 1]
df_5_filtered = df_5_filtered[df_5_filtered['Lifetime'] > 1]
# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df_10_filtered['Lifetime'], df_10_filtered['Average Neighbors'], color='blue', alpha=0.7, label='10')
plt.scatter(df_15_filtered['Lifetime'], df_15_filtered['Average Neighbors'], color='red', alpha=0.7, label='15')
plt.scatter(df_5_filtered['Lifetime'], df_5_filtered['Average Neighbors'], color='green', alpha=0.7, label='5')
#plt.xscale('log')
#plt.yscale('log')
#plt.scatter(df['Lifetime'], df['Neighbors'], color='red', alpha=0.7, label='Fish Data')
# Add titles and labels
plt.title('Fish Lifetime vs Average Neighbors', fontsize=16, fontweight='bold')
plt.xlabel('Lifetime (steps)', fontsize=14)
plt.ylabel('Average Neighbors', fontsize=14)

# Customize ticks and grid
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
