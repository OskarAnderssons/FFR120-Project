```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from CSV files
df_0 = pd.read_csv('Data_1312/Fish_eaten_log_0_1312.csv')
df_1 = pd.read_csv('Data_1312/Fish_eaten_log_1_1312.csv')
df_05 = pd.read_csv('Data_1312/Fish_eaten_log_05_1312.csv')
df_025 = pd.read_csv('Data_1312/Fish_eaten_log_025_1312.csv')
df_15 = pd.read_csv('Data_1312/Fish_eaten_log_15_1312.csv')
max_lifetime = df_1['Time Elapsed'].max()

slope_0 = np.polyfit(df_0['Time Elapsed'].div(10), df_0['Fish Eaten'], 1)[0]
slope_15 = np.polyfit(df_15['Time Elapsed'].div(10), df_15['Fish Eaten'], 1)[0]
slope_1 = np.polyfit(df_1['Time Elapsed'].div(10), df_1['Fish Eaten'], 1)[0]
slope_05 = np.polyfit(df_05['Time Elapsed'].div(10), df_05['Fish Eaten'], 1)[0]
slope_025 = np.polyfit(df_025['Time Elapsed'].div(10), df_025['Fish Eaten'], 1)[0]

print('Slope 1.5: ', slope_15, 'Slope 1: ', slope_1, 'Slope 0.5: ', slope_05, 'Slope 0.25: ', slope_025, 'Slope 0: ', slope_0)
# Create the scatter and line plot
plt.figure(figsize=(12, 6))

# Scatter plot of raw data
plt.plot(df_0['Time Elapsed'].div(10), df_0['Fish Eaten'], color='black', alpha=1, label='0.0s')
plt.plot(df_1['Time Elapsed'].div(10), df_1['Fish Eaten'], color='blue', alpha=1, label='1.0s')
plt.plot(df_05['Time Elapsed'].div(10), df_05['Fish Eaten'], color='red', alpha=1, label='0.5s')
plt.plot(df_025['Time Elapsed'].div(10), df_025['Fish Eaten'], color='green', alpha=1, label='0.25s')
plt.plot(df_15['Time Elapsed'].div(10), df_15['Fish Eaten'], color='orange', alpha=1, label='1.5s')
# Add titles and labels
plt.title('Fish Eaten vs Time', fontsize=22, fontweight='bold')
plt.xlabel('Time (seconds)', fontsize=16)
plt.ylabel('Total fish eaten', fontsize=16)

# Customize ticks and grid
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.3)
# plt.xscale('log')
# plt.yscale('log')

# Add legend
plt.legend(ncol=2, title='Negative Sensory Delay', title_fontsize=16, fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```
