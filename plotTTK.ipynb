```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
delayTTK = pd.read_csv('delay_TTK.csv')

# Extract data for plotting
sensory_delay = delayTTK['Sensory Delay']
average_ttk = delayTTK['TTK']

# Create evenly spaced indices for the bars
bar_positions = range(len(sensory_delay))

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(bar_positions, average_ttk, color='crimson', edgecolor='black')

# Add titles and labels
plt.title('Sensory Delay vs Average Time To Kill', fontsize=16)
plt.xlabel('Average Time To Kill (s)', fontsize=14)
plt.ylabel('Sensory Delay (s)', fontsize=14)


# Adjust y-axis to show sensory delay labels with equal spacing
plt.yticks(bar_positions, sensory_delay)

# Set x-axis limit to start at 200
plt.xlim(left=min(average_ttk)-10, right = max(average_ttk)+1)

plt.gca().invert_yaxis()
# Display the plot
plt.tight_layout()
plt.show()


```
